```python
%matplotlib inline
#import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes = True)
from sqlalchemy import create_engine


from datetime import datetime, timedelta
```

```python
#for calculating the distance of the orders

def haversine_np(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.    

    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km
```

```python
df_all = []
for x in path:
    df = pd.read_csv(x, usecols = ['ActivityId','ActivityType','Time', 'Latitude','Longitude', 'Speed'])
    df = df[df['ActivityType'] == 'Cycle' ]
    g = df.groupby(['ActivityId'])
    bike_starting = g.first().sort_index().reset_index().drop_duplicates()
    bike_ending = g.last().sort_values(by = ['Time']).reset_index().drop_duplicates()
    df = bike_starting.merge(bike_ending[['ActivityId', 'Time', 'Latitude', 'Longitude']], how = 'left', on = 'ActivityId')
    df_all.append(df)
bike_all = pd.concat(df_all, axis=0, ignore_index=True)
```

## Calculating distance per each trip

```python
df_all = []
for x in path:
    df = pd.read_csv(x, usecols = ['ActivityId','ActivityType','Time', 'Latitude','Longitude', 'Speed'])
    df = df[df['ActivityType'] == 'Cycle'].sort_values(['ActivityId', 'Time'])
    df = df[['ActivityId','Time', 'Latitude', 'Longitude']]
    df['Latitude_dest'] = df.groupby('ActivityId')['Latitude'].shift(-1)
    df['Longitude_dest'] = df.groupby('ActivityId')['Longitude'].shift(-1)
    df['distance'] = haversine_np(df['Longitude'],df['Latitude'],df['Longitude_dest'],df['Latitude_dest'])
    df = df.groupby('ActivityId').distance.sum().reset_index()
    df_all.append(df)
bike_distance = pd.concat(df_all, axis=0, ignore_index=True)
```
